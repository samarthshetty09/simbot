```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Replace 'path_to_your_survey_data.csv' with the path to your CSV file.
data_path = '/content/survey.csv'
survey_data = pd.read_csv(data_path)
survey_data.head()  # Display the first few rows to ensure it is loaded correctly.
```

```python
def extract_simulator_counts(data, column):
    """ Helper function to count occurrences of each simulator from a given column. """
    simulators = {}
    for entry in data[column].dropna():
        entries = [sim.strip() for sim in entry.split(',')]  # Split and strip to handle multiple entries
        for sim in entries:
            if sim in simulators:
                simulators[sim] += 1
            else:
                simulators[sim] = 1
    return simulators
```

```python
# Extracting simulator counts
tried_simulators = extract_simulator_counts(survey_data, 'What simulators have you tried?')

# Plotting
plt.figure(figsize=(9,7))
plt.pie(tried_simulators.values(), labels=tried_simulators.keys(), autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Tried Simulators')
plt.show()
```

```python
# Extracting simulator counts
working_simulators = extract_simulator_counts(survey_data, 'What simulators are working for you')

# Plotting
plt.figure(figsize=(9,7))
plt.pie(working_simulators.values(), labels=working_simulators.keys(), autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Working Simulators')
plt.show()
```
