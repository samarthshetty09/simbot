### Comparision of the 4 simulators for TurtleBot

```python
# Load and display the contents of the CSV file
file_path = 'data/comparision_metrics.csv'
data = pd.read_csv(file_path)

# Display all rows of the dataframe
data
```

```python
# Correcting the header and propagating simulator names
data.columns = ['Simulator', 'Iterations', 'CPU_Usage_Percentage', 'Memory_Usage_Percentage', 'FPS']
data['Simulator'] = data['Simulator'].fillna(method='ffill')

# Dropping rows with NaN values in 'Iterations' as they do not contain valid data
data = data.dropna(subset=['Iterations'])

# Display cleaned data
data
```

```python
# Calculating descriptive statistics for each metric grouped by simulator
stats = data.groupby('Simulator').agg({
    'CPU_Usage_Percentage': ['mean', 'std', 'median'],
    'Memory_Usage_Percentage': ['mean', 'std', 'median'],
    'FPS': ['mean', 'std', 'median']
}).round(2)

stats
```

### Steps Used for Analysis:
1. **Descriptive Statistics**: Compute mean, median, standard deviation for each metric across simulators.
2. **Visualizations**: Create plots to visually compare these metrics across simulators.
3. **Performance Analysis**: Discuss insights on which simulator might be more efficient or better suited for specific applications based on the observed data.

Here are the computed statistics for each simulator, organized by CPU usage, memory usage, and frames per second (FPS):

### CPU Usage
- **CoppaliaSim**: High average CPU usage (~288.68%) with a moderate standard deviation, indicating fluctuating usage.
- **Gazebo**: Much lower CPU usage (~35.17%) with minimal fluctuations.
- **IssacSim**: Lowest CPU usage (~5.46%) with slight variability.
- **WeBots**: Variable CPU usage (average ~6.45%), but has a high standard deviation due to some peaks.

### Memory Usage
- **CoppaliaSim** and **Gazebo** have very low memory usage (around 0.26% and 3.20%, respectively).
- **IssacSim**: Higher memory usage (~48.63%).
- **WeBots**: Significantly higher memory usage (~75.87%).

### FPS
- **CoppaliaSim**: Moderate FPS (~19.88) with low variability.
- **Gazebo**: Very low FPS (~0.29), indicating poor performance in terms of frame rate.
- **IssacSim**: Low FPS (~0.33), similar to Gazebo.
- **WeBots**: High FPS (~15.04) with low variability, showing good performance.

### Insights
- **CoppaliaSim** may be suitable for tasks requiring high computational power due to its high CPU usage.
- **Gazebo** and **IssacSim** are more efficient in terms of CPU and memory usage but offer lower FPS, which might limit their use in real-time applications.
- **WeBots** shows a balance between high FPS and higher CPU and memory usage, suitable for complex simulations requiring real-time rendering.

```python
# Setting up the plots
# Reassigning correct column names and cleaning up the data again
import matplotlib.pyplot as plt
data.columns = ['Simulator', 'Iterations', 'CPU_Usage_Percentage', 'Memory_Usage_Percentage', 'FPS']
data['Simulator'] = data['Simulator'].fillna(method='ffill')

# Dropping rows with NaN values in 'Iterations' as they do not contain valid data
data = data.dropna(subset=['Iterations'])

fig, axes = plt.subplots(3, 1, figsize=(10, 15))
fig.tight_layout(pad=6.0)

# Titles for plots
titles = ['CPU Usage (%)', 'Memory Usage (%)', 'Frames Per Second (FPS)']

# Data for plotting
metrics = ['CPU_Usage_Percentage', 'Memory_Usage_Percentage', 'FPS']

# Plotting each metric
for ax, metric, title in zip(axes, metrics, titles):
    for simulator in data['Simulator'].unique():
        subset = data[data['Simulator'] == simulator]
        ax.plot(subset['Iterations'], subset[metric], label=simulator)
    ax.set_title(title)
    ax.set_xlabel('Iterations')
    ax.set_ylabel(title)
    ax.legend()

plt.show()

```

Here are the plots comparing the CPU usage, memory usage, and frames per second (FPS) across different simulators:

1. **CPU Usage (%)**: You can see significant differences in the CPU usage profiles. CoppaliaSim consistently uses a lot more CPU compared to the others, indicating it might be processing more complex simulations or less optimized in terms of CPU usage.

2. **Memory Usage (%)**: The memory usage is relatively stable across iterations for each simulator, with WeBots showing significantly higher memory usage compared to others. This could impact performance on systems with limited memory resources.

3. **Frames Per Second (FPS)**: WeBots and CoppaliaSim show relatively higher and more stable FPS, indicating better performance in terms of rendering or processing frames. In contrast, Gazebo and IssacSim have much lower FPS, which might be problematic for real-time simulation tasks.

These visualizations help identify which simulators are more resource-intensive and which provide better real-time performance.
